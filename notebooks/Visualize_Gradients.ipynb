```python
import os
import torch
import numpy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def draw_histograms(directory):
    epochs = len(os.listdir(directory))
    fig=plt.figure(figsize=(12,6), dpi= 100)
    rows = 4
    for epoch in range(1, epochs + 1):
        ax=fig.add_subplot(rows, epochs//rows ,epoch)
        grad_norm = torch.load(os.path.join(directory, 'epoch_{}.pt'.format(epoch)), map_location=lambda storage, loc: storage)
        plt.hist(grad_norm.numpy(), bins=100)
        ax.set_title("Epoch {}".format(epoch))
    fig.tight_layout()  # Improves appearance a bit.
    plt.show()
```

```python
draw_histograms('../models/cifar_res_net_20e_2/')
```
